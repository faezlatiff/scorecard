### 3. Generate cumulative return series 
To compare performance of the funds. Generate the cumulative return series for the 3-year period (2019-2021) and for the most recent 1-year period.<BR><BR>
*Hint: use the pct_change and cumprod() methods*

```python
# your code here

# generate cum returns for 3y
stocks_df_adjclose_3y = stocks_df_price1['Adj Close']
cum_ret_series_3y =  (stocks_df_adjclose.pct_change()+1).cumprod() - 1

# subset of prices for 1 year 
stocks_df_price1_1y = stocks_df_price1['Adj Close'].loc['31/12/2020':]

# generate cum returns for 1y
cum_ret_series_1y = (stocks_df_price1_1y.pct_change()+1).cumprod() - 1

# check the first few rows of the cum return series.
print("3-year cum return series\n", cum_ret_series_3y.head())
print("1-year cum return series\n", cum_ret_series_1y.head())
```

## Annualised returns

### 5. Calculate annualised returns for the 1-year and 3-year period data sets.

Use the formulas to calculate the annualize rate of return (ARR): <br>
- period return = (ending price - starting price)/starting price
- ARR = (1+period return)^(1/N) - 1) where N = number of years
- The period return is the last row value in the cumulative return series
- N = 1 for the most recent 1-year period, N= 3 for the 3-year period

**DYI Q1: Calculate Annualised returns</font>**<BR>
What is the most recent 1-year annualized return for 
- SECT
- VFMF
- VFMO

What is the 3-year annualized return for 
- SECT
- VFMF
- VFMO
  
Round all values to 4 decimal places.

```python
# your code here

stocks_df_adjclose_1y = stocks_df_adjclose_3y.loc['31/12/2020':]
starting_price_1y = stocks_df_adjclose_1y.iloc[0]
ending_price_1y = stocks_df_adjclose_1y.iloc[-1]
period_return_1y = (ending_price_1y / starting_price_1y) / starting_price_1y
arr_1y = (1+period_return_1y) ** (1/3) - 1
print("1 year annualised return")
print(arr_1y.round(4))

starting_price = stocks_df_adjclose_3y.iloc[0]  # First row
ending_price = stocks_df_adjclose_3y.iloc[-1]   # Last row
period_return = (ending_price / starting_price) / starting_price
arr_3y = (1+period_return) ** (1/3) - 1
print("3-year annualised return")
print(arr_3y.round(4))

```

### 6. Calculate the relative peformance vs the benchmark

Round all values to 4 decimal places.
  

```python
# your code here

# hint: subtract the spy annualised return from the annualised return dataframe
spy_ann_ret_1y = arr_1y['SPY']
sect_ann_ret_1y = arr_1y['SECT']
vfmf_ann_ret_1y = arr_1y['VFMF']
vfmo_ann_ret_1y = arr_1y['VFMO']

text_1y = f"""
1-year Relative Performance
SECT: {(sect_ann_ret_1y - spy_ann_ret_1y).round(4)}
VFMF: {(vfmf_ann_ret_1y - spy_ann_ret_1y).round(4)}
VFMO: {(vfmo_ann_ret_1y - spy_ann_ret_1y).round(4)}
"""

spy_ann_ret_3y = arr_3y['SPY']
sect_ann_ret_3y = arr_3y['SECT']
vfmf_ann_ret_3y = arr_3y['VFMF']
vfmo_ann_ret_3y = arr_3y['VFMO']
text_3y = f"""
3-year Relative Performance
SECT: {(sect_ann_ret_3y - spy_ann_ret_3y).round(4)}
VFMF: {(vfmf_ann_ret_3y - spy_ann_ret_3y).round(4)}
VFMO: {(vfmo_ann_ret_3y - spy_ann_ret_3y).round(4)}
"""

# check the type and value of spy_ann_ret_1y before using
print(text_1y)
print(text_3y)
```
